# EDA & Demo Pipeline

This notebook demonstrates basic EDA and runs the demo pipeline on `data/dummy/sample_surveys.csv`.

NOTE: This notebook uses dummy data for demonstration. Do not paste real PHI here or export PHI from ULEAD.

```python
import pandas as pd
from src.utils import load_csv
from src.data_preprocessing import combine_prompt_response, filter_blank

df = load_csv('../data/dummy/sample_surveys.csv')
df = combine_prompt_response(df)
df = filter_blank(df)
df.head()
```

```python
# Run sentiment demo (this will download demo models)
from src.sentiment_model import SentimentClassifier
sc = SentimentClassifier()
preds = sc.predict(df['text'].tolist())
df = df.join(preds)
df
```

```python
# Simple bar chart of sentiment distribution
import matplotlib.pyplot as plt
counts = df['label'].value_counts()
counts.plot(kind='bar')
plt.title('Sentiment distribution (demo)')
plt.show()
```
